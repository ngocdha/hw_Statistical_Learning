### 1. For X ~ Gamma(6, .3)
 &nbsp; &nbsp;(a) create a plot of the pdf with appropriate axis labels, and a title.
   
 &nbsp; &nbsp;(b) create a plot of the CDF with appropriate axis labels, and a title.
   
 &nbsp; &nbsp;(c) compute P(X $\le$ 17.2).
   
 &nbsp; &nbsp;(d) compute P(15 $\le$ X $\le$ 20).

```r
# 1.(a)
curve(dgamma(x, 6, .3), from=0, to=60, ylab = "density", 
      main = expression(paste("pdf of Gamma model (", alpha, "=6, ", beta, "=.3)")))
abline(0,0)
```

```r
# 1.(b)
curve(pgamma(x, 6, .3), from=0, to=60, ylab = "density", 
      main = expression(paste("CDF of Gamma model (", alpha, "=6, ", beta, "=.3)")))
abline(0,0)
```

### 2. For X ~ Bin(100, .75)
 &nbsp; &nbsp;(a) create a plot of the pdf with appropriate axis labels, and a title.
   
 &nbsp; &nbsp;(b) create a plot of the CDF with appropriate axis labels, and a title.
   
 &nbsp; &nbsp;(c) compute P(X = 77).
   
 &nbsp; &nbsp;(d) compute P(X $\ge$ 72).

```r
# 2.(a)
x <- seq(50, 100, 1)
plot(x, dbinom(x, 100, .75), type ="h", ylab = "Probability",
     main = "pdf of Binomial model (n=100, p=.75)")
```

```r
# 2.(b)
x <- seq(50, 100, 1)
plot(x, pbinom(x, 100, .75), type ="h", ylab = "Probability",
     main = "CDF of Binomial function (n=100, p=.75)")
```

```r
# 2.(c)
dbinom(77, 100, .75)
```

```r
# 2.(d)
1 - pbinom(71, 100, .75)
```

### 4. Let X<sub>1</sub> ~ Gamma(shape=3.5, rate=2) and X<sub>2</sub> ~ Gamma(shape=4, rate=1.2)
 &nbsp; &nbsp;(a) plot the pdfs of X<sub>1</sub> and X<sub>2</sub> on the same plot in different colors with a legend.
   
 &nbsp; &nbsp;(b) plot the CDFs of X<sub>1</sub> and X<sub>2</sub> on the same plot in different colors with a legend.
   
 &nbsp; &nbsp;(c) Determine the median of each distribution
   
 &nbsp; &nbsp;(d) Determine P(2 < X < 10) for each distribution

```r
# 4.(a)
curve(dgamma(x, 3.5, 2), from=0, to=11, col="red", ylab = "density",
      main = "pdfs of Gamma(3.5, 2) and Gamma(4, 1.2)")
curve(dgamma(x, 4, 1.2), add=T)
abline(0,0)
legend(6, 0.45, legend=c("Gamma(3.5, 2)","Gamma(4, 1.2)"),lty=c(1,1),col=c("red","black"))
```

```r
# 4.(b)
curve(pgamma(x, 3.5, 2), from=0, to=11, col="red", ylab = "density",
      main = "CDFs of Gamma(3.5, 2) and Gamma(4, 1.2)")
curve(pgamma(x, 4, 1.2), add=T)
abline(0,0)
legend(6, 0.6, legend=c("Gamma(3.5, 2)","Gamma(4, 1.2)"),lty=c(1,1),col=c("red","black"))
```

```r
# 4.(c)
qgamma(0.5, 3.5, 2)  # median of gamma(3.5, 2)
qgamma(0.5, 4, 1.2)  # median of gamma(4, 1.2)
```

```r
# 4.(d)
pgamma(10, 3.5, 2) - pgamma(2, 3.5, 2)  # P(2 < X < 10) of gamma(3.5, 2)
pgamma(10, 4, 1.2) - pgamma(2, 4, 1.2)  # P(2 < X < 10) of gamma(4, 1.2)
```

### 5. Let X<sub>1</sub> ~ LNorm(mu=2, sigma=1.5) and X<sub>2</sub> ~ Gamma(shape=3.5, rate=0.7)
 &nbsp; &nbsp;(a) plot the pdfs of X<sub>1</sub> and X<sub>2</sub> on the same plot in different colors with a legend.

 &nbsp; &nbsp;(b) Add the means to the x-axis of the plot in part (a). Use text label the means.
   
 &nbsp; &nbsp;(c) plot the CDFs of X<sub>1</sub> and X<sub>2</sub> on the same plot in different colors with a legend.
   
 &nbsp; &nbsp;(d) Add a horizontal line at height y = 0.5 to the CDF plot. Where the line intersects the curve is the median of the distribution.

```r
# 5.(a)
curve(dlnorm(x, 2, 1.5), from=0, to=30, col="red", ylab = "density", ylim = c(0, 0.2),
      main = "pdfs of LNorm(2, 1.5) and Gamma(3.5, 0.7)")
curve(dgamma(x, 3.5, 0.7), col = "blue", add=T)
abline(0,0)
legend(10, 0.17, legend=c("LNorm(2, 1.5)","Gamma(3.5, 0.7)"),lty=c(1,1),col=c("red","blue"))

# 5.(b)
mean_lnorm = exp(2+((1.5)^2)/2)
mean_gamma = 3.5/0.7
points(mean_lnorm, 0, col="red", pch=19)
text(mean_lnorm, 0.02, "Mean of LNorm", col= "red")
points(mean_gamma, 0, col="blue", pch=19)
text(mean_gamma, 0.02, "Mean of Gamma", col= "blue")
```

```r
# 5.(c)
curve(plnorm(x, 2, 1.5), from=0, to=50, col="red", ylab = "density", ylim = c(0,1),
      main = "CDFs of LNorm(2, 1.5) and Gamma(3.5, 0.7)")
curve(pgamma(x, 3.5, 0.7), col = "blue", add=T)
abline(0,0)
legend(30, 0.7, legend=c("LNorm(2, 1.5)","Gamma(3.5, 0.7)"),lty=c(1,1),col=c("red","blue"))

# 5.(d)
abline(h=0.5)
text(11, 0.52, "median")
```
