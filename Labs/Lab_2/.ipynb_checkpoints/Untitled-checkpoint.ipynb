```r
library(readxl)
water_data <- read_excel("water_data.xlsx")
```

```r
attach(water_data)
```

```r
flow_data <- water_data[(Month == 6)&(Day == 10),]
```

```r
hist(flow_data$Flow, prob=T, main = "Flow data for June 10", xlab = "Flow(cfs)")
lines(density(flow_data$Flow))
```

-	The histogram seems to be from a Gamma model, with a long tail on the right.
-	Min value is 550; max value = 13400

```r
boxplot(flow_data$Flow, range = 1.2, horizontal=T, main = "Boxplot of Flow data for June 10", xlab = "Flow (cfs)", col = 5)
```

- There are 5 outliers, but no extreme outliers

```r
IQR(flow_data$Flow)
```

```r
summary(flow_data$Flow)
```
