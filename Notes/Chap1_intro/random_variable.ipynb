$\sigma^2 = E[(X-\mu)^2]$ = average value of $(X-\mu)^2$

$\sigma^2$ = variance = avg squared distance from $\mu$

$\sigma$ = standard deviation = $\sqrt(\sigma^2)$ = typical deviation from $\mu$

$\sigma^2$ has unit of $x^2$; $\sigma$ has unit of $x$; we typically use $\sigma$
