# Uniform Model

```r
# Plot the pdf of uniform over (10,50)
curve(dunif(x,10,50),from=10,to=50,ylim=c(0,0.03),ylab="density",col="orange",lwd=2)
abline(0,0,col="red")
```

```r
# Plot the CDF
curve(punif(x,10,50),from=10,to=50,ylab="density",col="forestgreen",lwd=2)
abline(0,0,col="red")
```

```r
qunif(c(0.05,0.95),10,50)
```

# Gamma Model

```r
curve(dgamma(x,2,.25), from=0, to=30, ylab = "density",ylim=c(0,.25))
curve(dgamma(x,1,.25),0,30,col="red",lty=2,add=T)
curve(dgamma(x,3,.25),0,30,col="blue",lty=3,add=T)
abline(0,0)
legend(15,.24,legend=c("Gamma(2,0.25)","Gamma(1,0.25)","Gamma(3,0.25)"),lty=c(1,2,3),col=c("black","red","blue"))
```

```r
curve(dexp(x,.1),0,50,ylab="density")
curve(dexp(x,.2),0,50,add=T,col="midnightblue",lty=4)
```

# Normal Model

```r
qnorm(c(1:10)/10,.8,.15)
```

```r
qnorm(.5,.8,.15)
```

```r
curve(pnorm(x,.8,.15),0,1.4,ylab="Probability")
axis(2,c(0:10)/10)
abline(h=.25)
```

```r
curve(dbeta(x,4,2),0,1,col="red")
curve(dbeta(x,.5,2),0,1,add=T)
curve(dbeta(x,.5,.5),0,1,add=T,col="blue")
```

```r
curve(dnorm(x,.5,.1),0,1)
curve(dbeta(x,12,12),0,1,add=T,col="red")
```

```r
curve(dlnorm(x,.5,4),0,1000)
curve(dlnorm(x,1,4),0,1000,add=T)
curve(dlnorm(x,1,.44),0,1000,add=T)
```
