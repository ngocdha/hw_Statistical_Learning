```r
hist(drat, prob=T, breaks=c(0:8)*.3+2.6, 
     xlab = "rear axle ratio", main = "Histogram of rear axle ratio")
points(drat,c(1:32)*0,pch="+")
abline(0,0)
lines(density(drat),col="red",lwd=1.5)
```

```r
hist(drat, prob=T, breaks=c(0:8)*.3+2.6, axes=F, 
     xlab = "rear axle ratio", main = "Histogram of rear axle ratio")
abline(0,0)
lines(density(drat),col="red",lwd=1.5)
```
